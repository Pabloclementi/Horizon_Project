```python
import pandas as pd

def dividir_direccion(df, columna):
    # Divide la columna 'columna' en partes
    df2 = df[columna].str.split(', ', n=3, expand=True)
    df2.columns = ['nombre_local', 'direccion', 'ciudad', 'codigo_postal_cod_ciudad']

    # Divide la columna 'codigo_postal' de df2 en partes adicionales
    df3 = df2['codigo_postal_cod_ciudad'].str.split(' ', n=1, expand=True)
    df3.columns = ['codigo_ciudad', 'codigo_postal']

    # Concatena df2 y df3 a lo largo de las columnas
    df_concatenado = pd.concat([df2, df3], axis=1)
    df_concatenado = pd.concat([df_concatenado, df],axis=1)
    return df_concatenado


df_resultado = dividir_direccion(df_metadata, 'address')

# Muestra el DataFrame resultante
df_resultado
```

```python
palabras_clave = [
    'restaurant', 'coffee', 'beer', 'food', 'wine', 'taco', 'burger', 
    'pizza', 'sushi', 'bar', 'bistro', 'cafe', 'diner', 'pub', 
    'bakery', 'brewery', 'grill', 'steakhouse', 'buffet', 'canteen',
    'brunch', 'cuisine', 'dessert', 'eatery', 'fast food', 'gastropub',
    'dinner', 'breakfast', 'lunch', 'sandwich', 'pasta', 'noodle', 
    'ice cream', 'smoothie', 'juice', 'food truck', 'food court', 'veggie','vegetarian', 'vegan', 'plant-based', 'veggie', 'tofu', 
    'tempeh', 'seitan', 'quinoa', 'lentils', 'beans', 
    'salad', 'greens', 'smoothie bowl', 'raw food', 'non-dairy', 
    'meatless', 'vegetables', 'fruit', 'herbivore', 'nutritional yeast'
]


# Función para buscar palabras clave en una lista de categorías
def es_gastronomico(categorias, palabras_clave):
    if isinstance(categorias, list):  # Ver si es una lista
        categorias_str = ' '.join(categorias)  # Convierte la lista en un string
        for palabra in palabras_clave:
            if palabra in categorias_str:
                return True
    return False

df_resultado['es_gastronomico'] = df_resultado['category'].apply(es_gastronomico, palabras_clave=palabras_clave)

# Filtra el DataFrame para incluir solo los negocios del rubro gastronómico
df_gastronomico = df_resultado[df_resultado['es_gastronomico']]

# Muestra el DataFrame resultante
df_gastronomico
```

```python
def filtrar_por_palabras_clave(df, columna, palabras_clave):
    def es_gastronomico(categorias, palabras_clave):
        if isinstance(categorias, list):
            categorias_str = ' '.join(categorias)
            for palabra in palabras_clave:
                if palabra in categorias_str:
                    return True
        return False

    df['es_gastronomico'] = df[columna].apply(es_gastronomico, palabras_clave=palabras_clave)
    return df[df['es_gastronomico']]

# Uso de la función con el DataFrame df_resultado y la columna 'category'
df_gastronomico = filtrar_por_palabras_clave(df_resultado, 'category', palabras_clave)

# Muestra el DataFrame resultante
print(df_gastronomico)
```
